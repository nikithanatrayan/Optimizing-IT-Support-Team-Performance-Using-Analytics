```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\user\Desktop\customer_support_tickets.csv")
```

```python
print(df.head())
print(df.shape)
print(df.info())
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop("Customer Name", axis=1, errors="ignore")
df = df.drop("Customer Email", axis=1, errors="ignore")
df = df.drop("Ticket Description", axis=1, errors="ignore")
df = df.drop("Ticket Status", axis=1, errors="ignore")
df = df.drop("Resolution", axis=1, errors="ignore")
df = df.drop("Time to Resolution", axis=1, errors="ignore")
df = df.drop("Customer Satisfaction Rating", axis=1, errors="ignore")
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
df.columns
```

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()          
        .str.replace(r'\s+', ' ', regex=True)  
    )
```

```python
df.to_csv("cleaned_customer_support_tickets.csv", index=False)
```
